```python
# Plot the transformed images (2 global + 4 local views)

import random
from matplotlib import pyplot as plt
from torchvision.datasets import CIFAR10

from transforms import MultiviewTransform, common_transform

transform = MultiviewTransform(nglobal=2, nlocal=4)
dataset = CIFAR10(root='./data', train=False, transform=common_transform, download=True)

im, label = dataset[random.randint(0, len(dataset)-1)]
im = transform(im)

fig, axs = plt.subplots(1, 6, figsize=(12, 4))
axs[0].imshow(im[0][:3].permute(1, 2, 0))
for i in range(5):
    axs[i+1].imshow(im[i+1][:3].permute(1, 2, 0))
plt.show();
```
